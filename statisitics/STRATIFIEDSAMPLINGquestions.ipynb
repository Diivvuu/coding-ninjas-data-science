# Stratified Sampling

```python
import pandas as pd
```

```python
df = pd.read_csv("startup_funding.csv")
df = df[(df['CityLocation'] == "Bangalore") |
        (df['CityLocation'] == "bangalore") | 
        (df['CityLocation'] == "New Delhi") |
        (df['CityLocation'] == "Delhi") |
        (df['CityLocation'] == "Mumbai") |
        (df['CityLocation'] == "Gurgaon")|
        (df['CityLocation'] == "Noida") |
        (df['CityLocation'] == "Pune") | 
        (df['CityLocation'] == "Hyderabad")]
```

```python
df['CityLocation'].replace("bangalore", "Bangalore")
df['CityLocation'].replace("Delhi", "New Delhi")
df.head()
```

```python
df_bangalore = df[(df['CityLocation'] == "Bangalore")]
df_ncr = df[(df['CityLocation'] == "New Delhi") | (df['CityLocation'] == "Gurgaon") | (df['CityLocation'] == "Noida")]
df_mumbai = df[(df['CityLocation'] == "Mumbai")]
df_pune = df[(df['CityLocation'] == "Pune")]
df_hyderabad = df[(df['CityLocation'] == "Hyderabad")]
```

```python
population = df
amount = population['AmountInUSD'].str.replace(',', '')
amount = pd.to_numeric(amount)
population_max = amount.max()
```

```python
s = []
bangalore_amount = df_bangalore['AmountInUSD'].str.replace(',', '').dropna()
bangalore_amount = pd.to_numeric(bangalore_amount)
s.append(bangalore_amount.max())

ncr_amount = df_ncr['AmountInUSD'].str.replace(',', '').dropna()
ncr_amount = pd.to_numeric(ncr_amount)
s.append(ncr_amount.max())

mumbai_amount = df_mumbai['AmountInUSD'].str.replace(',', '').dropna()
mumbai_amount = pd.to_numeric(mumbai_amount)
s.append(mumbai_amount.max())

pune_amount = df_pune['AmountInUSD'].str.replace(',', '').dropna()
pune_amount = pd.to_numeric(pune_amount)
s.append(pune_amount.max())

hyderabad_amount = df_hyderabad['AmountInUSD'].str.replace(',', '').dropna()
hyderabad_amount = pd.to_numeric(hyderabad_amount)
s.append(hyderabad_amount.max())

samples_max = max(s)
```

```python
error = population_max - samples_max
error
```

# Proportional Stratified Sampling

```python
import pandas as pd
df = pd.read_csv("startup_funding.csv")
df['CityLocation'].replace("bangalore", "Bangalore", inplace=True)
df['CityLocation'].replace("Delhi", "New Delhi", inplace = True)
df = df[(df['CityLocation'] == "Bangalore") | 
        (df['CityLocation'] == "New Delhi") |
        (df['CityLocation'] == "Gurgaon") |
        (df['CityLocation'] == "Noida") |
        (df['CityLocation'] == "Mumbai") | 
        (df['CityLocation'] == "Pune") | 
        (df["CityLocation"] == "Hyderabad")]
df
```

```python
population = df
population = population['AmountInUSD'].str.replace(',', '')
population = pd.to_numeric(population)
max_of_population = max(population)
max_of_population
```

```python
df_bangalore = df[(df['CityLocation'] == "Bangalore")]
df_ncr = df[(df['CityLocation'] == "New Delhi") | (df['CityLocation'] == "Noida") | (df['CityLocation'] == "Gurgaon")]
df_mumbai = df[(df['CityLocation'] == "Mumbai")]
df_pune = df[(df['CityLocation'] == "Pune")]
df_hyderabad = df[(df['CityLocation'] == "Hyderabad")]
```

```python
population_size = len(df)
population_size
bangalore_sample_size = round(20*((len(df_bangalore))/population_size))
ncr_sample_size = round(20*((len(df_ncr))/population_size))
mumbai_sample_size = round(20*((len(df_mumbai))/population_size))
pune_sample_size = round(20*((len(df_pune))/population_size))
hyderabad_sample_size = round(20*((len(df_hyderabad))/population_size))

# total = len(df_bangalore) + len(df_ncr) + len(df_mumbai) + len(df_pune) + len(df_hyderabad)
# print(total)
```

```python
print(bangalore_sample_size)
print(ncr_sample_size)
print(mumbai_sample_size)
print(pune_sample_size)
print(hyderabad_sample_size)
```

```python
bangalore_sample = df_bangalore.sample(bangalore_sample_size)
ncr_sample = df_ncr.sample(ncr_sample_size)
mumbai_sample = df_mumbai.sample(mumbai_sample_size)
pune_sample = df_pune.sample(pune_sample_size)
hyderabad_sample = df_hyderabad.sample(hyderabad_sample_size)
```

```python
bangalore_sample = bangalore_sample['AmountInUSD'].str.replace(',','')
bangalore_sample = pd.to_numeric(bangalore_sample)

ncr_sample = ncr_sample['AmountInUSD'].str.replace(',','')
ncr_sample = pd.to_numeric(ncr_sample)

mumbai_sample = mumbai_sample['AmountInUSD'].str.replace(',','')
mumbai_sample = pd.to_numeric(mumbai_sample)

pune_sample = pune_sample['AmountInUSD'].str.replace(',','')
pune_sample = pd.to_numeric(pune_sample)

hyderabad_sample = hyderabad_sample['AmountInUSD'].str.replace(',','')
hyderabad_sample = pd.to_numeric(hyderabad_sample)
```

```python
s = []
s.append(bangalore_sample.max())
s.append(ncr_sample.max())
s.append(mumbai_sample.max())
s.append(pune_sample.max())
s.append(hyderabad_sample.max())
samples_max = max(s)
```

```python
sampling_error = population_max - samples_max
sampling_error
```
