```python
import csv

with open('Datasets/year2017.csv') as file_obj :
    file_data = csv.reader(file_obj, skipinitialspace = True)
    
    file_list = list(file_data)

killed = []
for row in file_list[1:] :
    val = row[9]
    if val != '' :
        killed.append(float(val))
    
print(int(sum(killed)))
```
