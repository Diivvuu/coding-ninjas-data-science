```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
title = []
for i in df_train['Name'].values:
    if 'Master' in i:
        title.append(0)
    elif 'Miss' in i:
        title.append(1)
    elif 'Mr' in i:
        title.append(2)
    elif 'Mrs' in i:
        title.append(3)
    else:
        title.append(4)
```

```python
df_train['Embarked'].value_counts()
```

```python
df_train.loc[df_train['Embarked'] == 'S', ['Embarked']] = 1
df_train.loc[df_train['Embarked'] == 'C', ['Embarked']] = 2
df_train.loc[df_train['Embarked'] == 'Q', ['Embarked']] = 3
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])
```

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mode()[0])
```

```python
df_train.isna().sum()
```

```python
df_train.loc[df_train['Sex'] == 'female', ['Sex']] = int(0) 
df_train.loc[df_train['Sex'] == 'male', ['Sex']] = int(1)
```

```python
df_train['Cabin'] = df_train['Cabin'].fillna('N')
```

```python
cabin = []
for i in df_train['Cabin'].values:
    if 'A' in i:
        cabin.append(1)
    elif 'B' in i:
        cabin.append(2)
    elif 'C' in i:
        cabin.append(3)
    elif 'D' in i:
        cabin.append(4)
    elif 'E' in i:
        cabin.append(5)
    elif 'F' in i:
        cabin.append(6)
    elif 'G' in i:
        cabin.append(7)
    elif 'T' in i:
        cabin.append(8)
    elif 'N' in i:
        cabin.append(0)
```

```python
Cabin = pd.DataFrame(cabin, columns=['Cabin'])
```

```python
Cabin.value_counts()
```

```python
df_train['Ticket'].value_counts()
```

```python
ticket = []
for i in df_train['Ticket'].values:
    if i.split()[-1] == 'LINE' : 
        ticket.append(0)
    else:
        ticket.append(int(i.split()[-1]))

Ticket = pd.DataFrame(ticket, columns = ['Ticket'])
```

```python
df_train['Ticket'].values
```

```python
df_train.drop('Name', inplace=True, axis=1)
df_train.drop('Ticket', inplace=True, axis=1)
df_train.drop('Cabin', inplace=True, axis=1)
```

```python
df_train=pd.concat([pd.DataFrame(title, columns=['Title']), pd.DataFrame(Cabin, columns=['Cabin']), pd.DataFrame(Ticket, columns=['Ticket']), df_train], axis=1)
```

```python
df_train
```

```python
df_train['Ticket'] = df_train['Ticket']/max(df_train['Ticket'])
df_train['Age'] = df_train['Age']/max(df_train['Age'])
df_train['Fare'] = df_train['Fare']/max(df_train['Fare'])
df_train['Pclass'] = df_train['Pclass']/max(df_train['Pclass'])
```

```python
df_train.iloc[:, :10]
```

```python
df_train.dtypes
```

```python
df_train = df_train.astype('float')
```

```python
x_train = df_train.iloc[:, :10].values
y_train = df_train.iloc[:, 10:].values
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
title_test = []
for i in df_test['Name'].values:
    if 'Master' in i:
        title_test.append(0)
    elif 'Miss' in i:
        title_test.append(1)
    elif 'Mr' in i:
        title_test.append(2)
    elif 'Mrs' in i:
        title_test.append(3)
    else:
        title_test.append(4)

df_test.loc[df_test['Embarked'] == 'S', ['Embarked']] = 1
df_test.loc[df_test['Embarked'] == 'Q', ['Embarked']] = 2
df_test.loc[df_test['Embarked'] == 'C', ['Embarked']] = 3
df_test['Embarked'] = df_test['Embarked'].fillna(df_test['Embarked'].mode()[0])
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mode()[0])
df_test.loc[df_test['Sex'] == 'female', ['Sex']] = int(0)
df_test.loc[df_test['Sex'] == 'male', ['Sex']] = int(1)
df_test['Cabin'] = df_test['Cabin'].fillna('N')

cabin_test = []
for i in df_test['Cabin'].values:
    if 'A' in i:
        cabin_test.append(1)
    if 'B' in i:
        cabin_test.append(2)
    if'C' in i:
        cabin_test.append(3)
    if 'D' in i:
        cabin_test.append(4)
    if 'E' in i:
        cabin_test.append(5)
    if 'F' in i:
        cabin_test.append(6)
    if 'G' in i:
        cabin_test.append(7)
    if 'T' in i:
        cabin_test.append(8)
    if 'N' in i:
        cabin_test.append(0)

Cabin_test = pd.DataFrame(cabin_test, columns = ['Cabin'])

ticket_test = []
for i in df_test['Ticket'].values:
    if i.split()[-1] == 'LINE':
        ticket_test.append(0)
    else:
        ticket_test.append(int(i.split()[-1]))

Ticket_Test = pd.DataFrame(ticket_test, columns = ['Ticket'])

df_test.drop('Name', inplace = True, axis = 1)
df_test.drop('Ticket', inplace = True, axis = 1)
df_test.drop('Cabin', inplace = True, axis = 1)

df_test = pd.concat([pd.DataFrame(title_test, columns = ['Title']), pd.DataFrame(Cabin_test, columns = ['Cabin']), pd.DataFrame(Ticket_Test, columns = ['Ticket']), df_test], axis = 1)

df_test['Ticket'] = df_test['Ticket']/max(df_test['Ticket'])
df_test['Age'] = df_test['Age']/max(df_test['Age'])
df_test['Fare'] = df_test['Fare']/max(df_test['Fare'])
df_test['Pclass'] = df_test['Pclass']/max(df_test['Pclass'])

df_test=df_test.astype(float)
x_test = df_test.values
```
