# Terror Casualty Attack

```python
import numpy as np
import csv
file_obj = open('terrorismData.csv', encoding="UTF-8")
data = csv.DictReader(file_obj, skipinitialspace=True)

#List of only those columns which are to be used in output
wounded = []
killed = []
city = []
group = []
for row in data:
    if row['State'] == "Jammu and Kashmir":
        if row['Year'] == '1999':
            if row['Month'] == '5' or row['Month'] == '6' or row['Month'] == '7':
                if "Unknown" not in row['City']:
                    if "Unknown" not in row['Group']:
                        killed.append(row['Killed'])
                        wounded.append(row['Wounded'])
                        city.append(row['City'])
                        group.append(row['Group'])

# Converting to np arrays
killed = np.array(killed)
wounded = np.array(wounded)
city = np.array(city)
group = np.array(group)

#Removing null values by 0.0
killed[killed == ""] = "0.0"
wounded[wounded == ""] = "0.0"


# Convering to float from str
killed = np.array(killed, dtype=float)
wounded = np.array(wounded, dtype=float)

# calculating casuality
casualty = killed + wounded
max_casualty = (int)(casualty.max())

# get index of max casuality
max_casualty_index = casualty.argmax()
print(max_casualty, city[max_casualty_index], group[max_casualty_index])
```

# Terror Casualty Red

# Terror City Casualty

```python
import numpy as np
import csv
file_obj = open('terrorismData.csv', encoding="UTF-8")
data = csv.DictReader(file_obj)

city = []
killed = []
wounded = []
for row in data:
    if row['Country'] == 'India' and "Unknown" not in row['City']:
        city.append(row['City'])
        killed.append(row['Killed'])
        wounded.append(row['Wounded'])


city = np.array(city)
killed = np.array(killed)
wounded = np.array(wounded)
killed[killed == ""] = "0.0"
wounded[wounded == ""] = "0.0"
killed = np.array(killed, dtype=float)
wounded = np.array(wounded, dtype=float)
casualty = np.array(killed + wounded, dtype = 'int')

citydic = {}
for i in range(len(city)):
    if city[i] in citydic:
        citydic[city[i]] += casualty[i]
    else:
        citydic[city[i]] = casualty[i]

for j in range(5):
    count = 0
    city = ''
    for i in citydic:
        if citydic[i] > count:
            count = citydic[i]
            city = i
    print(city, count)
    del citydic[city]
```
