# mpg dataset

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.relplot(x = 'mpg', y = 'horsepower', data = df)
plt.show()
print("Yes")
```

# Origin with Maximum data points

```python
sns.relplot(x = 'mpg', y = 'horsepower', data = df, hue = 'origin')
plt.show()
print("USA")
```
