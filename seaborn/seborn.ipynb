# Visualising Statistical relationship


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'smoker', style = 'time')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, col = 'smoker', hue = 'time')
plt.show()
```

# Visualizing Statistical Relations

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('flights')
df.head()
```

#### We used line plt becauase scatter would not be suitable, visualization would not be clear in scatter plot


```python
sns.relplot(x = 'year', y = 'passengers', data = df, kind = 'line')
plt.show()
```

```python
sns.relplot(x = 'year', y = 'passengers', data = df, kind = 'line', hue = 'month')
plt.show()
```
