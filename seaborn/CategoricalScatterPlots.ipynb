# Categorical Scatter PLots

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

### stripplot

```python
sns.catplot(x = 'sex', y='total_bill', data=df)
plt.show()
```

### jitter parameter

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, jitter = 1, hue = 'smoker')
plt.show()
```

### swarm plot

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = 'swarm', hue = 'smoker')
plt.show()
```

## Box PLot

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df,  kind = 'box', hue = 'smoker')
plt.show()
```

# Boxen Plot
#### very similar to Box Plot

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df,  kind = 'boxen', hue = 'smoker')
plt.show()
```

## Violin Plot

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df,  kind = 'violin')
plt.show()
```

```python
sns.catplot(x = 'species', y = 'sepal_width', data = df, kind = 'violin')
plt.show()
```
