```python
import numpy as np
from sklearn import preprocessing
from sklearn import model_selection
```

```python
training = np.loadtxt('0000000000002419_training_ccpp_x_y_train.csv', delimiter = ',')
testing = np.loadtxt('0000000000002419_test_ccpp_x_test.csv', delimiter = ',')
```

```python
def step_gradient(X,Y, m, learning_rate):
    m_slope = np.zeros(len(X[0]))
    for i in range(len(X)):
        x = X[i]
        y = Y[i]
        for j in range(len(x)):
            m_slope[j] += (-2/len(X))*(y - sum(m*x))*x[j]
    new_m = m - (learning_rate*m_slope)
    return new_m
```

```python
def cost(m, x, y):
    cost = 0
    for i in range(len(x)):
        cost+= (1/len(x))*((y[i]-sum(m*x[i]))**2)
    print(cost)
```

```python
def gd(x, y, learning_rate, num_iterations):
    m = np.zeros(len(x[0]))
    for i in range(num_iterations):
        m = step_gradient(x, y, m, learning_rate)
        print("itr= ", i, "cost= ", end = ' ')
        cost(m, x, y) 
    return m
```

```python
def gradient_descent(x, y):
    num_iterations = 200
    learning_rate = 0.11
    x = np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m = gd(x, y, learning_rate, num_iterations)
    return m
```

```python
x = training[:, :-1]
y = training[:, -1]

sq = []
for i in range(len(x[0])):
    for j in range(i, len(x[0])):
        sq.append(x[:, i]*x[:, j])
sq = np.array(sq)

for i in sq:
    x =np.append(x, i.reshape(-1, 1), axis = 1)

scaler = preprocessing.StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
m = gradient_descent(x, y)
```

```python
sq = []
for i in range(len(testing[0])):
    for j in range(i, len(testing[0])):
        sq.append(testing[:, i]*testing[:, j])
sq = np.array(sq)
for i in sq:
    testing=np.append(testing, i.reshape(-1, 1), axis=1)

testing1=scaler.transform(testing)
x_test=np.append(testing1, np.ones(len(testing1)).reshape(-1, 1), axis=1)
ans=[]
for i in x_test:
    ans.append(sum(i*m))
for i in ans:
    print(i)
ans=np.array(ans)
np.savetxt(X=ans,fname='dataa_power_plant.csv' , delimiter=',', fmt='%.5f')
```
