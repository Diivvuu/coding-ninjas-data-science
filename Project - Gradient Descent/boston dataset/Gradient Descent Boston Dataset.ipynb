```python
import numpy as np
import sklearn.preprocessing as ppg
import sklearn.model_selection as ms
from sklearn.linear_model import LinearRegression as LR
from sklearn.ensemble import GradientBoostingRegressor as GBR
training = np.loadtxt('0000000000002417_training_boston_x_y_train.csv', delimiter=',')
testing = np.loadtxt('0000000000002417_test_boston_x_test.csv', delimiter=',')
```

```python
x_train = training[:, :-1]
y_train = training[:, -1]
x_test = testing
scaler = ppg.StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = GBR()
model.fit(x_train, y_train)
y_test = model.predict(x_test)
np.savetxt("Divu_Scaler_predicted.csv", y_test)
```

```python
model.score(x_train, y_train)
```
