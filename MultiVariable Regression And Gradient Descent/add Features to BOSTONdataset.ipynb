### Adding all 2 degree features to BOSTON dataset

```python
import pandas as pd
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
X = boston.data
Y = boston.target
df = pd.DataFrame(X)
columns = boston.feature_names
df.columns =  columns
df
```

```python
for i in range(len(boston.feature_names)):
    for j in range(i, len(boston.feature_names)):
        index = columns[i] + '_' + columns[j]
        df[index] = df[columns[i]]*df[columns[j]]
X = df.values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
alg = LinearRegression()
alg.fit(X_train, Y_train)
```

```python
training_Score = alg.score(X_train, Y_train)
Testing_Score = alg.score(X_test, Y_test)
print("Training Score : ", training_Score)
print("Testing Score : ", Testing_Score)
```
