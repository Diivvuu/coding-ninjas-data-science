# PCA for compression

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('https://files.codingninjas.in/naruto-7582.png')
print(type(img), img.shape)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
temp = img.copy()
img_shape = img.shape
temp = temp.reshape((img_shape[0]*img_shape[1], 4)) #rehsaping the image
```

```python
pca = PCA(n_components=1, whiten=True)
xtr = pca.fit(temp)
xtr = pca.transform(temp)
xtr.shape
```

```python
xtr = xtr.reshape(img_shape[0], img_shape[1])
xtr.shape
```

```python
fig = plt.figure(figsize = (16, 16))
ax = fig.add_subplot(2, 2, 1)
ax.imshow(img, cmap = plt.cm.bone)
ax = fig.add_subplot(2, 2, 2)
ax.imshow(xtr, cmap = plt.cm.bone)
plt.show()
```
