# Microsoft Bubble Graph

```python
import numpy as np
import matplotlib.pyplot as plt
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
employees = np.array(employees)
year = np.array(year)
revenue = np.array(revenue)
count = np.arange(len(year))
plt.scatter(year, revenue, s = employees, c = count)
plt.axis([2004, 2019, 38, 120])
for i in range(len(year)):
    plt.text(year[i] + 0.1, revenue[i] + 0.1, employees[i])
plt.grid()
index1 = np.where(year == 2008) 
index2 = np.where(year == 2015)
print(year[index1[0][0]],revenue[index1[0][0]], employees[index1[0][0]])
print(year[index2[0][0]],revenue[index2[0][0]], employees[index2[0][0]])
plt.show()
```

# Laptop Sales

```python
import matplotlib.pyplot as plt
import numpy as np
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
sold = np.array(sold)
perc = np.true_divide(sold, sold.sum(axis=0))*100
plt.pie(sold, labels = company, autopct='%.2f')
plt.show()
for i in range(len(perc)):
    print(company[i], format(perc[i],'.1f'))
```

# Plot Job_year

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("amazon_jobs_dataset.csv")
year = data.Posting_date.str[-4:]
sorted_year = sorted(year)
year_dict={}
for i in sorted_year:
    if i in year_dict :
        year_dict[i] += 1
    else:
        year_dict[i] = 1
plt.plot(year_dict.keys(), year_dict.values())
for i in year_dict.keys():
    print(i, year_dict[i])
```

# Plot Month_Job

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import collections
data = pd.read_csv("amazon_jobs_dataset.csv")
month = data['Posting_date'].str.split(' ').str[0]
month = np.array(month)
month_dict = {}
for i in month:
    if i in month_dict:
        month_dict[i] += 1
    else:
        month_dict[i] = 1
month_dict = collections.OrderedDict(sorted(month_dict.items()))  
plt.bar(month_dict.keys(), month_dict.values())
plt.xticks(rotation = 40)
plt.show()
for i in month_dict.keys():
    print(i, month_dict[i])
```

# Plot city_job

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('amazon_jobs_dataset.csv')
df_location=df['IN' == df['location'].str.split(', ').str[0]]
cities = df_location.location.str.split(', ').str[2]
freq = cities.value_counts()
plt.axis("Equal")
plt.pie(freq, labels = freq.index)
plt.show()
perx = np.divide(freq, freq.sum(axis = 0))*100
for i in freq.index:
    print(i, format(perx[i], '.2f'))
```

# Plot Job_Java

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import csv
with open('amazon_jobs_dataset.csv', encoding = "UTF-8") as file_obj:
    data = csv.DictReader(file_obj)    
    year = {}
    for row in data:
        if 'java' in row["BASIC QUALIFICATIONS"] or 'Java' in row["BASIC QUALIFICATIONS"]:
            saal = row['Posting_date'].split()[2]
            if saal in year:
                year[saal] += 1
            else:
                year[saal] = 1
# year = sorted(year.keys)
for i in sorted(year.keys()):
    print(i, year[i])
year_keys = sorted(year.keys())
year_values = sorted(year.values())
plt.scatter(year_keys, year_values)
plt.show()
```
